```python
import os
from os import listdir
from os.path import isfile, join
import pandas as pd
import numpy as np

user = 'BIEN'
if user=='BIEN':
    path_file = '/Users/NhatMinh/Desktop/ML/Stage_data/all_data_without2408versionV0.csv'
elif user == 'HUYEN':
    path_file = '/Users/lehuyen/Dropbox/Stage_data/all_data_without2408versionV0.csv'
else:
    path_file = None
    
def load_and_save_to_file():
    
    if user=='BIEN':
        directory='/Users/NhatMinh/Desktop/ML/Stage_data/insa01'
    elif user =='HUYEN':
        directory='/Users/lehuyen/Dropbox/Stage_data/insa01' 
    else:
        directory = None

    onlyfiles = [f for f in listdir(directory) if isfile(join(directory, f)) and '2408' not in f]
    index = 0
    list_all_df = []
    for  file in onlyfiles:
        if 'csv' in file:
            if index ==0:
                path = os.path.join(directory, file)
                df_all = pd.read_csv(path,  sep='\n', header=0, delimiter=";",decimal=',')
                list_all_df.append(df_all)
            else:
                path = os.path.join(directory, file)
                this_df = pd.read_csv(path,  sep='\n', header=0, delimiter=";",decimal=',')
                df_all = df_all.append(this_df)
                list_all_df.append(this_df)
            index +=1
    df_all['TimeT'] = df_all.Time.map(lambda x: x.split('+')[0][:-3][5:] )
    df_all = df_all.set_index('TimeT')
    df_all = df_all.drop(['Time'], axis=1)
    df_all.columns == ['CPU','MEM','REC', 'TRAN','Status']
    df_all.to_csv(path_file)

try: 
    insa = pd.read_csv(path_file)
except FileNotFoundError:
    load_and_save_to_file()
    insa = pd.read_csv(path_file)
insa = insa.set_index('TimeT')
print(insa.shape)
insa.head()
```
